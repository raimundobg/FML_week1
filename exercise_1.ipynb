```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

#importing data frames for analysis
df1 = pd.read_csv('steps.csv', delimiter =';')

df1.head()
```

```python
#importing data frames for analysis
df2 = pd.read_csv('survey.csv')
df2.head()
```

```python
df1['Q_steps'] = df1.drop('id', axis=1).sum(axis=1)
df1['Q_used_days'] = df1.drop('id', axis=1).count(axis=1)
df1.head()
```

```python
#joining datframes
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head()
```

1. Getting the media of steps per user is useful to create a visualization according to all users and their single average, give insights about how the data is distributed

```python
# getting the media of steps for all users in all days
df = df.reset_index(drop=True)
#getting rid of columns that are not steps
df_mean = df.drop(columns=['weight', 'city', 'age', 'gender', 'hh_size', 'education', 'education_1' , 'height', 'Q_steps', 'Q_used_days'])
df_mean['average'] = df_mean.mean(axis = 1) 
df_mean.head()
```

```python
# scatterplot for spotting outliers
sns.scatterplot(x='id', y='average', data=df_mean)
plt.xlabel('id')
plt.ylabel('media of steps for total days')
plt.show()
```

```python
#outliers above 20.000 will be removed
df_mean["average"] = df_mean["average"]
def outlier(x):
    if(x > 20000): 
        return float('NaN')
    else: 
        return x
df_mean["average"] = df_mean["average"].apply(outlier)
```

2. without outlier > than 20.000 the distribution shows range of values from 0 to 20.000 as average for the total steps of users

```python
#visualization after removing outliers
df_cleaned = df_mean
sns.scatterplot(x='id', y='average', data=df_cleaned)
plt.xlabel('id')
plt.ylabel('media of steps for total days')
plt.show()
```
